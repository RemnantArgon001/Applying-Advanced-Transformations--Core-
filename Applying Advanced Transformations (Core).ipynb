Measurments needs to be split height and weight should be 2 columns and the values be the data in that column.

```python
Left['Measurements'].head(2)
```

```python
to_replace = [',']
for char in to_replace:
    Left['Measurements'] = Left['Measurements'].str.replace(char,':',regex=False)
```

```python
Left['Measurements'].str.split(':',expand=True)
```

```python
Left[['0','height (cm)','2','weight (kg)']] = Left['Measurements'].str.split(':',expand=True)
Left.head(2)
```

```python
Left = Left.drop(columns=['Measurements', '0', '2'])
Left.head(2)
```

```python
to_replace = ["'", ' cm', ' kg', '}',' ']
for char in to_replace:
    Left['height (cm)'] = Left['height (cm)'].str.replace(char,'',regex=False)
```

```python
to_replace = ["'", 'cm', 'kg', '}',' ']
for char in to_replace:
    Left['weight (kg)'] = Left['weight (kg)'].str.replace(char,'',regex=False)
```

Starting right df we need to address the powers column will need to make column for each power by one-hot-encoding right

```python
Right.loc[2,'Powers']
```

```python
Right['Powers'].str.split(',',expand=False)
```

```python
Right['power_split'] = Right['Powers'].str.split(",")
```

```python
Right['power_split'].head()
```

```python
exploded = Right.explode('power_split', ignore_index=False)
exploded[['hero_names','Powers','power_split']].head(5)
```

```python
cols_to_make = exploded['power_split'].dropna().unique()
```

```python
for col in cols_to_make:
    Right[col] = Right['Powers'].str.contains(col)
Right.head()
```

```python
merged= pd.merge(Left, Right, left_on='hero', right_on='hero_names')
merged.head()
```
